```python
import numpy as np
from geon.data.pointcloud import PointCloudData
from geon.data.document import Document
import numpy as np

doc = Document()



# pcd 1
points = np.ones((43,3),float)
doc.add_data(PointCloudData(points)) 

# pcd 2
points = np.ones((57,3),float)
pcd2 = PointCloudData(points)

doc.add_data(pcd2)








```

```python
from geon.data.pointcloud import FieldType, SemanticSchema, SemanticClass
schema = SemanticSchema()
schema.add_semantic_class('botanik')
schema.add_semantic_class('corocodile')

pcd2.add_field(field_type=FieldType.SEMANTIC, schema=schema)
```

```python
doc.save_hdf5('./test3.h5')
```
